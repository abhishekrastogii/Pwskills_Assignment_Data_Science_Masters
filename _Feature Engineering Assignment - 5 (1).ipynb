Q1: Pearson Correlation Coefficient

Calculate the Pearson correlation coefficient between study time and exam scores.

python

import numpy as np
from scipy.stats import pearsonr

study_time = np.array([1, 2, 3, 4, 5])  # hours studied
exam_scores = np.array([50, 60, 70, 80, 90])  # exam scores

correlation, _ = pearsonr(study_time, exam_scores)
print(f"Pearson Correlation Coefficient: {correlation}")

if correlation > 0:
    print("Positive linear relationship.")
elif correlation < 0:
    print("Negative linear relationship.")
else:
    print("No linear relationship.")

    
Q2: Spearman's Rank Correlation

Calculate Spearman's rank correlation between sleep hours and job satisfaction.

python

from scipy.stats import spearmanr

sleep_hours = np.array([7, 6, 5, 8, 9])  # hours of sleep
job_satisfaction = np.array([5, 4, 3, 8, 9])  # satisfaction levels

spearman_correlation, _ = spearmanr(sleep_hours, job_satisfaction)
print(f"Spearman's Rank Correlation: {spearman_correlation}")

if spearman_correlation > 0:
    print("Positive monotonic relationship.")
elif spearman_correlation < 0:
    print("Negative monotonic relationship.")
else:
    print("No monotonic relationship.")
